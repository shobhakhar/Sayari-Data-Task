```python
# extracting json file that has information about company with company names
url = "https://firststop.sos.nd.gov/api/Records/businesssearch"

payload = json.dumps({
  "SEARCH_VALUE": "x",
  "STARTS_WITH_YN": "true",
  "ACTIVE_ONLY_YN": True
})
headers = {
  'authority': 'firststop.sos.nd.gov',
  'accept': '*/*',
  'accept-language': 'en-US,en;q=0.9',
  'authorization': 'undefined',
  'content-type': 'application/json',
  'cookie': 'ASP.NET_SessionId=za5t3mpcb3dgqw3wrk4usgw2',
  'origin': 'https://firststop.sos.nd.gov',
  'referer': 'https://firststop.sos.nd.gov/search/business',
  'sec-ch-ua': '"Chromium";v="104", " Not A;Brand";v="99", "Google Chrome";v="104"',
  'sec-ch-ua-mobile': '?0',
  'sec-ch-ua-platform': '"Windows"',
  'sec-fetch-dest': 'empty',
  'sec-fetch-mode': 'cors',
  'sec-fetch-site': 'same-origin',
  'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
}
response = requests.request("POST", url, headers=headers, data=payload)
```

```python
data = json.loads(response.text)
```

```python
keys = data["rows"].keys()
```

```python
# Storing comapany names in a list.
company_name =[]
for key in keys:
    company_name.append(data["rows"][key]["TITLE"][0])
```

```python
# extract owners or agent name for each company
url = "https://firststop.sos.nd.gov/api/FilingDetail/business/"
company_owners = []
company_agents =[]
for key in keys:
    url1 = url+key+"/false"

    payload={}
    headers = {
      'authority': 'firststop.sos.nd.gov',
      'accept': '*/*',
      'accept-language': 'en-US,en;q=0.9',
      'authorization': 'undefined',
      'cookie': 'ASP.NET_SessionId=za5t3mpcb3dgqw3wrk4usgw2',
      'referer': 'https://firststop.sos.nd.gov/search/business',
      'sec-ch-ua': '"Chromium";v="104", " Not A;Brand";v="99", "Google Chrome";v="104"',
      'sec-ch-ua-mobile': '?0',
      'sec-ch-ua-platform': '"Windows"',
      'sec-fetch-dest': 'empty',
      'sec-fetch-mode': 'cors',
      'sec-fetch-site': 'same-origin',
      'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
    }

    response = requests.request("GET", url1, headers=headers, data=payload)
    data1 = json.loads(response.text)
    if len(data1["DRAWER_DETAIL_LIST"])==10:
        company_owners.append(data1['DRAWER_DETAIL_LIST'][2]["VALUE"])
        company_agents.append("none")
    else:
        company_owners.append("none")
        company_agents.append(data1['DRAWER_DETAIL_LIST'][-1]["VALUE"])
```

```python
len(company_agents),len(company_owners)
```

```python
# dictionary of lists 
dict = {"Company_ids":keys,'Company_name': company_name, 'Company_owners': company_owners, 'Company_agents': company_agents} 
    
df = pd.DataFrame(dict)
```

```python
df.head(10)
```

```python
# remove characters after "\n" including "\n"
df["Company_agents"] = df["Company_agents"].str.replace("(\n).*","")
df["Company_owners"] = df["Company_owners"].str.replace("(\n).*","")
```

```python
df.tail(20)
```

```python
df1 = df.copy()
df1.info()
```

```python
df1["Company_agents_owners"] = np.where(df1["Company_owners"] == "none", df1["Company_agents"], df1["Company_owners"])
```

```python
# drop unwanted columns
df1.drop(columns=['Company_owners',"Company_agents","Company_ids"], inplace=True)
```

```python
df1.head(25)
```

```python
df2 = df1.drop(df1.index[:20]) # remove companies that do not start with X.
```

```python
df2.head()
```

```python
# change characters to uppercase
df2["Company_name"] = df2["Company_name"].str.upper()
df2["Company_agents_owners"] = df2["Company_agents_owners"].str.upper()
```

```python
len(df2["Company_name"].unique()),len(df2["Company_agents_owners"].unique())
```

```python
#Now, let us get the unique list of company_names and Company_agents_owners from our dataframe
Unique_list = list(set(df2.Company_name).union(set(df2.Company_agents_owners))) 
```

```python
# Initialize and add nodes
sdt_net = Network(height='750px', width='100%', bgcolor='#222222', font_color='white',notebook=True)
sdt_net.add_nodes(Unique_list)
```

```python
# list of edges
my_sdt_list = df2.values.tolist()
```

```python
#Now, let us add edges and viusualize our network
sdt_net.add_edges(my_sdt_list)
sdt_net.show("my_sdt_map.html")
```
